```python
rand_6x3 = np.random.rand(6, 3)
transposed = rand_6x3.T
print(rand_6x3)
print(transposed)
```

```python
n, p, m = 2, 3, 4
A = np.random.rand(n, p)
B = np.random.rand(p, m)
dot_product = A @ B
print(A)
print(B)
print(dot_product)
```
